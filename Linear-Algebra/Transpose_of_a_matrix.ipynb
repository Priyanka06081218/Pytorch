```python
# Input matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

def transpose(matrix):
    transpose = []

    for col in range(len(matrix[0])):
        row = []

        for r in matrix:
            row.append(r[col])

        transpose.append(row)

    return transpose


result = transpose(matrix)

print("Original Matrix:")
for row in matrix:
    print(row)

print("\nTranspose Matrix:")
for row in result:
    print(row)
```
