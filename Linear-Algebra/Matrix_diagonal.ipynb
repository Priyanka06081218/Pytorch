```python
import numpy as np

def make_diagonal(v):
    result = []

    for i in range(len(v)):
        row = []

        for j in range(len(v)):
            if i == j:
                row.append(v[i])
            else:
                row.append(0)

        result.append(row)

    return np.array(result)

# Input
n = int(input("Enter the number of elements: "))

v = []
for i in range(n):
    value = int(input(f"Enter element {i+1}: "))
    v.append(value)

# Output
diagonal_matrix = make_diagonal(v)

print("\nDiagonal Matrix:")
print(diagonal_matrix)
```

### Using numpy make_diagonal

```python
import numpy as np

def make_diagonal(v):
    return np.diag(v)

# Input
n = int(input("Enter the number of elements: "))

v = []
for i in range(n):
    value = int(input(f"Enter element {i+1}: "))
    v.append(value)

# Output
diagonal_matrix = make_diagonal(v)

print("\nDiagonal Matrix:")
print(diagonal_matrix)
```
