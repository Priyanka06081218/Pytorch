```python
def matrix_trace(A):
    result = 0

    for i in range(len(A)):
        result += A[i][i]

    return result


# Input
n = int(input("Enter the size of the square matrix (n): "))

matrix = []

print("Enter the matrix row by row:")

for i in range(n):
    row = list(map(float, input(f"Row {i+1}: ").split()))
    matrix.append(row)

# Output
trace = matrix_trace(matrix)

print("\nMatrix:")
for row in matrix:
    print(row)

print("\nTrace =", trace)
```
